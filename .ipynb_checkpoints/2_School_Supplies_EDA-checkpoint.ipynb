## 1- Date_Time Feature

```python
print('The data set covers the period from {}, to {}'.format(df.date_time.min(), df.date_time.max()))
```

```python
# Create a function by passing a datetime object extracting year, month and day from x

def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    
    return year, month, day
```

```python
# Extract year, month and day from InvoiceDate

review_year, review_month, review_day = get_date_int(df, 'date_time')
```

```python
# Create an InvoiceYear, InvoiceMonth, InvoiceDay column

df['review_year']= review_year
df['review_month']= review_month
```

```python
# Check on which year there are more sales.
plt.figure(figsize = (6,6))
plt.subplot(211)
sns.countplot(x=df.review_year)
plt.xlabel('Years')
plt.ylabel("# of reviews")

plt.subplot(212)
sns.countplot(x=df.review_month)
plt.xlabel('Months')
plt.ylabel("# of reviews")

plt.show()
```

### Top 10 Words in the Corpus

```python
top_N = 10
                
cleaned_text = [y for x in df.normalized_text for y in x.split()
                if len(y)>2 and y not in 'not, nan, one, two, three, four, five, six, seven, eight, nine']

word_dist = nltk.FreqDist(cleaned_text)
rslt = pd.DataFrame(word_dist.most_common(top_N),
                    columns=['Word', 'Frequency'])

plt.figure(figsize=(8,6))
sns.set()
ax = sns.barplot(x="Word",y="Frequency", data=rslt.head(10))
```

In the top ten words, four of them are very positive (great, well, like, good)

## Good Rating & Bad Rating Words

```python
# Create a new data frame with clean text and rating class
df2 = df[["normalized_text", "rating_class_num"]]
df2.head(5)
```

```python
# Initialize the countervectorizer
countVec = CountVectorizer(binary=True)
```

```python
# Fit the 'clean_text' to countvectorizer
countVec.fit(df["normalized_text"])
```

```python
# Transform the matrix
transformed_matrix = countVec.transform(df["normalized_text"])
```

```python
# Convert matrix to array
transformed_matrix.toarray()
```

```python
# Extracting the feature names
names = countVec.get_feature_names()
```

```python
# Adding a 'rating' column from previous dataframe's rating value
df3 = pd.DataFrame(transformed_matrix.toarray(), columns=names)
df3['rating'] = df['rating_class_num']
```

```python
# Create lists for forming a dataframe summary
avg_ratings = [] 
for name in names:    
    avg_rating = df3[df3[name]== 1]['rating'].mean()
    avg_ratings.append(avg_rating)
```

```python
avg_rating
```

```python
# Create lists for forming a dataframe summary
rating_counts = []
for name in names:    
    rating_count = df3[df3[name]== 1]['rating'].count()
    rating_counts.append(rating_count)  
```

```python
# Create lists for forming a dataframe summary
feature_names = []
for name in names:    
    feature_names.append(name )
```

```python
# Create a new dataframe from words, average ratings, and rating counts
df_summary = pd.DataFrame({'feature_name':feature_names, 'avg_ratings': avg_ratings, 'rating_counts':rating_counts})
```

```python
# Let's see the new dataframe

df_summary = df_summary[df_summary.feature_name != 'rating']
df_summary.head(10)
```

```python
# Words that are commonly used in the reviews which have good ratings
df_good = df_summary.query("rating_counts > 100").sort_values(by='avg_ratings', ascending=False)[:100]
```

```python
df_good.head(10)
```

```python
wc_good = dict(zip(df_good['feature_name'].tolist(), df_good['avg_ratings'].tolist()))
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate_from_frequencies(wc_good)
 
# plot the WordCloud image                       
plt.figure(figsize = (8, 18), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
# Words that are commonly used in the reviews which have bad ratings
df_bad = df_summary.query("rating_counts > 100").sort_values(by= 'avg_ratings', ascending=True)[:100]
```

```python
df_bad.head(10)
```

```python
wc_bad = dict(zip(df_bad['feature_name'].tolist(), df_bad['avg_ratings'].tolist()))
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate_from_frequencies(wc_bad)
 
# plot the WordCloud image                       
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
